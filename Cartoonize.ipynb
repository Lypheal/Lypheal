```python
import cv2
import numpy as np
# Only uncomment these 2 lines if using in google colab
# from google.colab import drive
# drive.mount("/content/drive")
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/drive/MyDrive/input.png") #edit this path 

# 1) Edges
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 5)
edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)

# 2) Color
color = cv2.bilateralFilter(img, 9, 300, 300)

# 3) Cartoon
sketch = cv2.bitwise_and(color, color, mask=edges)

path = "/content/drive/MyDrive/output.png" #edit this aswell
cv2_imshow(sketch)
cv2.imwrite(path,sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
